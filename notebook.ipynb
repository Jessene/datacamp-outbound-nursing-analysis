clean_and_validate_calls_data <- function(calls_data) {
  # Remove duplicate rows
  calls_data <- unique(calls_data)
  # Column-specific validation and cleaning
  cleaned_data <- calls_data %>%
    mutate(
      # Convert to character and trim whitespace (should already be alright)
      patient_id = as.character(patient_id),
      patient_id = str_trim(patient_id),
      # Ensure uppercase, remove whitespace
      screening_type = str_to_upper(str_trim(screening_type)),
      screening_type = if_else(
        screening_type %in% c("BCS", "COL", "EED", "CBP", "OMW"),
        screening_type,
        NA_character_),
      # Convert to numeric (suppressing NAs by coercion), validate
      screening_completed_ind = suppressWarnings(as.numeric(str_trim(screening_completed_ind))),
      screening_completed_ind = if_else(
        screening_completed_ind %in% c(0, 1) | is.na(screening_completed_ind),
        screening_completed_ind,
        NA_real_),
      # Parse and validate date against the present (today is 2025-12-01)
      latest_call_date = ymd(latest_call_date),
      latest_call_date = if_else(
        latest_call_date <= Sys.Date(), 
        latest_call_date, 
        as.Date(NA)
      ),
      # Convert to numeric
      reached_ind = as.numeric(str_trim(reached_ind)),
      reached_ind = if_else(
        reached_ind %in% c(0, 1) | is.na(reached_ind), 
        reached_ind, 
        NA_real_
      ),
      # Parse and validate 
      screening_date = ymd(screening_date),
      screening_date = if_else(
        screening_date <= Sys.Date(), 
        screening_date, 
        as.Date(NA)
      )
    )
  return(cleaned_data)
}

# Safe T-test function because OMW has only 2 for not reached
safe_t_test <- function(data, formula) {
  group_var_name <- as.character(formula)[[3]]
  counts <- table(data[[group_var_name]])
  if (length(counts) == 2 && all(counts >= 2)) {
    return(t.test(formula, data = data)$p.value)
  } else {
    return(as.numeric(NA))
  }
}

                                 family = binomial(link = "logit"))

model_summary <- summary(model_compliance_vs_count)
p_value <- model_summary$coefficients["total_assigned", "Pr(>|z|)"]
z_value <- model_summary$coefficients["total_assigned", "z value"]
estimate <- model_summary$coefficients["total_assigned", "Estimate"]

stats_subtitle <- glue("Coefficient: {round(estimate, 3)}, z-value: {round(z_value, 2)}, p-value: {format.pval(p_value, digits = 3, eps = 0.001)}")
